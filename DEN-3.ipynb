# Iris Flower Dataset Analysis

In this notebook, we will explore and visualize the Iris flower dataset, a well-known dataset in the field of machine learning. The dataset contains measurements of sepal length, sepal width, petal length, and petal width for three species of iris flowers: Iris setosa, Iris versicolor, and Iris virginica. 

We'll use `pandas` for data manipulation and `matplotlib` and `seaborn` for data visualization.



## 1. Import Necessary Libraries
We will use pandas for data manipulation, matplotlib, and seaborn for visualization.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the Dataset

We'll load the Iris dataset from the UCI Machine Learning Repository and assign appropriate column names.


```python
# Load the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Species"]
iris_data = pd.read_csv(url, names=columns)

# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(iris_data.head())
```

## Summary Statistics

Let's explore the summary statistics of the dataset to understand the distributions of the numeric features.


```python
# Summary statistics of the dataset
print("\nSummary Statistics:")
print(iris_data.describe())
```

## Check for Missing Values

We will check if there are any missing values in the dataset.

```python
# Check for missing values
print("\nMissing Values:")
print(iris_data.isnull().sum())
```

## Data Visualization

Now, let's visualize the data to uncover patterns and relationships.


```python
# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# 1. Scatter plot of Sepal Length vs Sepal Width
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris_data, x='SepalLength', y='SepalWidth', hue='Species', style='Species', s=100)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

### Scatter Plot Analysis

The scatter plot shows the relationship between sepal length and sepal width, color-coded by species. This visualization helps us see how the species are distributed in this two-dimensional space.


```python
# 2. Pairplot of the dataset
sns.pairplot(iris_data, hue='Species')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

## Conclusion

In this analysis, we explored the Iris flower dataset through summary statistics and various visualizations. We identified relationships among features and differences between species, highlighting how sepal and petal measurements can be used to classify iris flowers. This notebook serves as a foundational step for further analysis and machine learning applications using this dataset.
